```python
__author__ = 'MinterS'

#Direct 2.0 CleanseMatch 

import json, requests

CREDENTIALS = {
    "production":{
        "uid":"P100000B2F115D3F0F34D8A9A0B7B238",
        "pwd":"Welcome@24",
    },
}

BASE_URL = "https://maxcvservices.dnb.com/V5.0/organizations?CountryISOAlpha2Code=US&SubjectName={Subject}&TerritoryName={State}&cleansematch=true"
AUTH_URL = "https://maxcvservices.dnb.com/rest/Authentication/V2.0"

#Returns the authorization token
def getAuth(uid, pwd):
    headers = {"x-dnb-user":uid, "x-dnb-pwd":pwd}

    result = requests.post(AUTH_URL, headers=headers)
    return result.headers["Authorization"]

prodAuth = getAuth(CREDENTIALS["production"]["uid"], CREDENTIALS["production"]["pwd"])

#Returns the response body
def sendRequest(Subject, State):
    global prodAuth
    url = BASE_URL
    url = url.replace("{Subject}", Subject)
    url = url.replace("{State}", State)
    #url = url.replace("{Country}", Country)
    headers = {"Authorization": prodAuth}
    #print url
    response = requests.get(url,headers=headers)
    result = json.loads(response.text)
    return result
```

```python
import csv
stations = list()
with open('Stations.tab', 'r') as f:
    reader = csv.reader(f, delimiter='\t')
    for row in reader:
        if row[0] != 'State':
            stations.append(row)
```
